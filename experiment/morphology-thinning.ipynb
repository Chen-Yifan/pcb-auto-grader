```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
import os


%matplotlib inline
```

```python
def compare_img(original, predicted):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('original')
    
    plt.subplot(1, 2, 2)
    plt.imshow(predicted)
    plt.title('predicted')
    
def check_erosion(path, iterations=6, show_img=True):
    img = cv2.imread(path)
    kernel = np.ones((5,5))
    erosion = cv2.erode(img,kernel,iterations = iterations)
    if show_img:
        compare_img(img, erosion)
    return erosion
```

```python
def compare_regulation_mask(path, iterations=6, show_img=True):
    img = cv2.imread(path)
    kernel = np.ones((5,5))
    erosion = cv2.erode(img,kernel,iterations = iterations)
    if show_img:
        compare_img(img, np.where(erosion[:,:,1] == 88, 1, 0) + np.where(erosion[:,:,1] == 89, 1, 0))
    return erosion

def compare_other_mask(path, iterations=6, show_img=True):
    img = cv2.imread(path)
    kernel = np.ones((5,5))
    erosion = cv2.erode(img,kernel,iterations = iterations)
    if show_img:
        compare_img(img, np.where(erosion[:,:,1] == 0, 1, 0))
    return erosion
```

```python
root = './data/processed_data/'
for f in sorted(os.listdir(root)):
    if 'copper' not in f and 'top' in f:
        compare_regulation_mask(root+f)
```

```python
root = './data/processed_data/'
for f in sorted(os.listdir(root)):
    if 'copper' not in f and 'top' in f:
        compare_other_mask(root+f)
```
