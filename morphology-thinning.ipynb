```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
import os


%matplotlib inline
```

```python
def compare_img(original, predicted):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('original')
    
    plt.subplot(1, 2, 2)
    plt.imshow(predicted)
    plt.title('predicted')
    
def check_erosion(path, iterations=6):
    img = cv2.imread(path)
    kernel = np.ones((5,5))
    erosion = cv2.erode(img,kernel,iterations = iterations)
    compare_img(img, erosion)
```

```python
root = './data/processed_data/'
for f in os.listdir(root):
    if 'top-copper' in f:
        print(f)
        check_erosion(root+f)
```

```python
root = './data/processed_data/'
for f in os.listdir(root):
    if 'copper' not in f and 'top' in f:
        print(f)
        check_erosion(root+f, iterations=3)
```
